```python
np.arange(0, len(vid_con))
```

```python
vid_con['t'] = np.arange(0, len(vid_con))
```

```python
def contour_plot(df, thresh):
    fig, ax = plt.subplots(figsize=(20, 6), layout='constrained')
    df1 = df[df['contours'] < thresh]
    df2 = df[df['contours'] >= thresh]
    ax.plot(df1['t'], df1['contours'], color = 'grey')
    ax.plot(df2['t'], df2['contours'], color = 'red')
    # df['thresh'] = np.where(df['contours'] > thresh, True, False)
    # f = sns.scatterplot(data=df, x = 't', y = 'contours', hue = 'thresh')
    # f._legend.remove()
    # return ax
```

```python
contour_plot(vid_con, 1550)
```

```python
vid_df = pd.DataFrame(data = np.loadtxt('/home/lab/Documents/all_vids/HD_vids/22_04_2023_CS/contours_22_04_2023_CS.txt'), columns = ['contours'])
vid_df['t'] = np.arange(0, len(vid_df))
```
